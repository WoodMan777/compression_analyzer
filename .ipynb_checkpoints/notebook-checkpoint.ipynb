```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from time import time
import bitstring as bs
```

```python
def getFastBinString(binf):
    res = np.array([], dtype = "int8")
    """#data = []
    
    for i in binf:
        data.append((i >> vecmul)&1)
        #np.concatenate((res,)"""
    
    vecmul = np.array([7,6,5,4,3,2,1,0])
    return np.concatenate(list(map(lambda x: (x >> vecmul)&1, binf)))

def getBinString(binf):
    res = ""
    vecmul = np.array([128,64,32,16,8,4,2,1])
    for i in range(len(binf)):
        tmp = ""
        for j in range(8):
            tmp += chr(48 + ((binf[i] >> (7-j))&1))
        res += tmp
    
    return res

def getSuperFastBinString(binf):
    return bs.BitArray(binf).bin
            
```

```python
def getMarkovMatrix(bits, blockSize):
    count = np.zeros((2**blockSize, 2**blockSize), dtype = "float64")
    
    padding = blockSize - len(bits)%(blockSize)
    bits += '0'*padding
    
    blockCount = len(bits) // blockSize
    
    for i in range(blockCount - 2):
        tmpa = bits[(blockSize)*i : (blockSize)*(i+1)]
        tmpb = bits[(blockSize)*(i+1) : (blockSize)*(i+2)]
        
        #print(f"tmpa: {tmpa}")
        #print(f"tmpb: {tmpb}")
        
        count[int(tmpa,2)][int(tmpb,2)] += 1
    
    return count / (blockCount)
```
